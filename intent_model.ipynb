```python


import json
import pandas as pd

# Load dataset
with open("dataset.json", "r", encoding="utf-8") as f: # Adjust the file path as needed
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

df.head()
```

```python
df.head(50)
```

```python
X = df["text"]
y = df["intent"]

print("Number of samples:", len(X))
print("Unique intents:", y.unique())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X_vectorized = vectorizer.fit_transform(X)

print("Feature shape:", X_vectorized.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(class_weight="balanced", max_iter=1000)

model.fit(X_vectorized, y)

print("Model retrained with class balancing.")
```

```python
# Test sentences
test_sentences = [
    "Ritu ko 2400 ka bill bana do",
    "Gupta Store ke liye 5000 ka invoice generate karo",
    "1800 ka bill create karo Pooja ke naam se",
    "Kal ka 900 ka bill bana do Neha ko",
    "Sharma ji ko 1200 ka invoice bhejna hai",
    "Ritu ko 2400 ka bill bana do",
    "Gupta Store ke liye 5000 ka invoice generate karo",
    "1800 ka bill create karo Pooja ke naam se",
    "Ritu ko 2400 ka bill bana do",
    "Gupta Store ke liye 5000 ka invoice generate karo",
    "1800 ka bill create karo Pooja ke naam se",
    "2000 ka maal diya Rahul ko",
    "900 ka bana do uske liye",
    "Usko 1100 ka saman gaya",
    "700 ka likh do",
    "Sharma ji ko 2500 ka entry kar do"
]

# Convert to vector
test_vector = vectorizer.transform(test_sentences)

# Predict
predictions = model.predict(test_vector)

for sentence, pred in zip(test_sentences, predictions):
    print(f"Sentence: {sentence}")
    print(f"Predicted Intent: {pred}")
    print("-" * 40)
```

```python
print(y.value_counts())
```

```python
import numpy as np

probs = model.predict_proba(test_vector)

for sentence, pred, prob in zip(test_sentences, predictions, probs):
    confidence = np.max(prob)
    print(f"Sentence: {sentence}")
    print(f"Predicted Intent: {pred}")
    print(f"Confidence: {confidence:.2f}")
    print("-" * 40)
```

```python
with open("dataset.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)

print(df["intent"].value_counts())
```

```python
X = df["text"]
y = df["intent"]

vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(X)

model = LogisticRegression(class_weight="balanced", max_iter=1000)
model.fit(X_vectorized, y)

print("3-class model trained successfully.")
```
